```python
for year in range(2000, 2002):

  # Imports
  mydf_temp_imports = comtradeapicall.getFinalData(subscription_key, # API key required for authentication.
                                              typeCode='C', # Type of data (e.g., 'C' = commodities).
                                              freqCode='A', # Frequency of data ('A' = annual, 'M' = monthly, etc.).
                                              clCode='HS',  # Classification system ('HS' = Harmonized System codes).
                                              period=year,     # Time period(s) requested (e.g., '2000' = year 2000).
                                              reporterCode=156, # Country code of the reporting country (e.g., 156 = China).
                                              cmdCode='AG2', # Commodity code(s) (e.g., 'AG2' = specific product group).
                                              flowCode='M', # Trade flow(s) ('M' = imports, 'X' = exports, both if comma-separated).
                                              partnerCode=None, # Country code for partner country (None = all).
                                              partner2Code=0,   # *Secondary* partner code (0 often = world aggregate).
                                              customsCode=None, # Customs regime code (e.g., imports under certain customs categories).
                                              motCode=None,     # Mode of transport code (e.g., sea, air, etc.).
                                              maxRecords=250000, # Maximum number of records to return (default 250,000); One API Call caps at 250000
                                              format_output='JSON', # Output format ('JSON', 'CSV', etc.).
                                              aggregateBy=None, # Field to aggregate data by (e.g., commodity, country).
                                              breakdownMode='classic', # Level of detail ('classic' = default breakdown).
                                              countOnly=None,   # If True, return only record counts instead of full data.
                                              includeDesc=True) # If True, include descriptive text (names of codes) in results.

  # Saving it to csv in the session for checking later
  mydf_temp_imports.to_csv(f'China_as_reporter_{year}_imports.csv')

  # Exports
  mydf_temp_exports = comtradeapicall.getFinalData(subscription_key, # API key required for authentication.
                                              typeCode='C', # Type of data (e.g., 'C' = commodities).
                                              freqCode='A', # Frequency of data ('A' = annual, 'M' = monthly, etc.).
                                              clCode='HS',  # Classification system ('HS' = Harmonized System codes).
                                              period=year,     # Time period(s) requested (e.g., '2000' = year 2000).
                                              reporterCode=156, # Country code of the reporting country (e.g., 156 = China).
                                              cmdCode='AG2', # Commodity code(s) (e.g., 'AG2' = specific product group).
                                              flowCode='X', # Trade flow(s) ('M' = imports, 'X' = exports, both if comma-separated).
                                              partnerCode=None, # Country code for partner country (can be None = all).
                                              partner2Code=0,   # Secondary partner code (0 often = world aggregate).
                                              customsCode=None, # Customs regime code (e.g., imports under certain customs categories).
                                              motCode=None,     # Mode of transport code (e.g., sea, air, etc.).
                                              maxRecords=250000, # Maximum number of records to return (default 250,000).
                                              format_output='JSON', # Output format ('JSON', 'CSV', etc.).
                                              aggregateBy=None, # Field to aggregate data by (e.g., commodity, country).
                                              breakdownMode='classic', # Level of detail ('classic' = default breakdown).
                                              countOnly=None,   # If True, return only record counts instead of full data.
                                              includeDesc=True) # If True, include descriptive text (names of codes) in results.

  # Saving it to csv in the session for checking later
  mydf_temp_exports.to_csv(f'China_as_reporter_{year}_exports.csv')

  # Combine imports + exports into one DataFrame
  mydf_all = pd.concat([mydf_temp_imports, mydf_temp_exports], ignore_index=True)

  # Concatenated file to be saved
  mydf_all.to_csv(f'China_as_reporter_{year}.csv')
```

### Using Async in getFinalData()

```python
# Submitting the async request and getting the Job ID
job = comtradeapicall.submitAsyncFinalDataRequest(subscription_key, # API key required for authentication.
                             typeCode='C', # Type of data (e.g., 'C' = commodities).
                             freqCode='A', # Frequency of data ('A' = annual, 'M' = monthly, etc.).
                             clCode='HS',  # Classification system ('HS' = Harmonized System codes).
                             period=2023,     # Time period(s) requested (e.g., '2000' = year 2000).
                             reporterCode=156, # Country code of the reporting country (e.g., 156 = China).
                             cmdCode='AG6', # Commodity code(s) (e.g., 'AG2' = specific product group).
                             flowCode='M,X', # Trade flow(s) ('M' = imports, 'X' = exports, both if comma-separated).
                             partnerCode=None, # Country code for partner country (None = all).
                             partner2Code=0,   # *Secondary* partner code (0 often = world aggregate).
                             customsCode=None, # Customs regime code (e.g., imports under certain customs categories).
                             motCode=None) # If True, include descriptive text (names of codes) in results.
```

```python
requestId = job['requestId']
```

```python
status = comtradeapicall.checkAsyncDataRequest(subscription_key, requestId)
```

```python
status['status']
```

```python
status['startTime'] = pd.to_datetime(status['startTime'])
status['endTime'] = pd.to_datetime(status['endTime'])
```

```python
# Time
print( status['endTime'] - status['startTime'] )
```

##### Using the status object to get the dataframe

```python
# Using the generated uri, i.e. status['uri'] to get the data

import zipfile
import io

# Check only if the request is complete
if status['status'][0] == 'Completed':

  file_uri = status['uri'][0]
  r = requests.get(file_uri)
  z = zipfile.ZipFile(io.BytesIO(r.content))
  csv_name = z.namelist()[0]
  df = pd.read_csv(z.open(csv_name), delimiter='\t')
  print(df.head())
```
